```python
import pandas as pd
```

```python
df=pd.read_csv('emails.csv')
```

```python
df.head()
```

```python
x=df.drop('Prediction',axis='columns').drop('Email No.',axis='columns')
```

```python
x
```

```python
df.head()
```

```python
y=df['Prediction']
```

```python
y
```

```python
from sklearn.neighbors import KNeighborsClassifier
#import sklearn as sn
from sklearn import metrics as mt
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test= train_test_split(x,y, test_size=0.20, random_state=10)
```

```python
x_train
```

```python
knnc=KNeighborsClassifier(n_neighbors=2)
```

```python
knnc.fit(x_train,y_train)
```

```python
output=knnc.predict(x_test)
```

```python
output
```

```python
mt.accuracy_score(output , y_test)
```

```python
knnc.score(x_test,y_test)
```

```python
mt.mean_squared_error(output,y_test)
```

```python
mt.mean_squared_error(output,y_test,squared= False)
```

```python
mt.confusion_matrix(output,y_test)
```

```python
from sklearn.svm import SVC
```

```python
svm= SVC()
```

```python
svm.fit(x_train,y_train)
```

```python
output2=svm.predict(x_test)
```

```python
output2
```

```python
svm.score(x_test,y_test)
```

```python
mt.accuracy_score(output2,y_test)
```

```python
mt.mean_squared_error(output2,y_test)
```

```python
mt.mean_squared_error(output2,y_test,squared=False)
```

```python
mt.confusion_matrix(output2,y_test)
```
