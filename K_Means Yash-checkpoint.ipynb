```python
from sklearn.cluster import KMeans
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("sales_data_sample.csv", encoding='Latin')
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df = df.drop(['TERRITORY' , 'POSTALCODE' , 'ADDRESSLINE2' ,'STATE'] , axis='columns')
```

```python
df.head()
```

```python
x=df['QUANTITYORDERED']
y=df['PRICEEACH']
```

```python

x.head()
```

```python
y.head()
```

```python
plt.scatter(x,y)
plt.xlabel('QUANTITYORDERED')
plt.ylabel('PRICEEACH($)')
```

```python
km = KMeans(n_clusters=2)
y_predicted = km.fit_predict(df[['QUANTITYORDERED','PRICEEACH']])
y_predicted
```

```python
df['cluster']=y_predicted
df
```

```python
km.cluster_centers_
```

```python
df1 = df[df.cluster==0]
df2 = df[df.cluster==1]

plt.scatter(df1.QUANTITYORDERED,df1['PRICEEACH'],color='green')
plt.scatter(df2.QUANTITYORDERED,df2['PRICEEACH'],color='red')

plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],color='purple',marker='*',label='centroid')
plt.xlabel('QUANTITYORDERED')
plt.ylabel('PRICEEACH')
plt.legend()
```

```python
sse = []
k_rng = range(1,10)
for k in k_rng:
    km = KMeans(n_clusters=k)
    km.fit(df[['QUANTITYORDERED','PRICEEACH']])
    sse.append(km.inertia_)
```

```python
plt.xlabel('K')
plt.ylabel('Sum of squared error')
plt.plot(k_rng,sse)
```
